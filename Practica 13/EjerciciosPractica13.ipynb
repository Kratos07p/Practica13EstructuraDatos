```python
'''Diseña una clase llamada Biblioteca que utilice una pila para gestionar libros prestados.
Implementa métodos para prestar, devolver y mostrar el estado de la pila. Crea una
instancia de la clase y realiza operaciones de préstamo y devolución con libros específicos
como "Cien años de soledad" o "Don Quijote de la Mancha". Imprime mensajes
informativos para cada operación y muestra el estado de la pila después de ciertas
acciones.'''

class Libro:
    def __init__(self, titulo, cantidad):
        self.titulo = titulo
        self.cantidad = cantidad

class Biblioteca:
    def __init__(self):
        self.catalogo = {}
        self.pila_prestamo = []

    def agregar_libro(self, libro):
        self.catalogo[libro.titulo] = libro

    def prestar_libro(self, titulo):
        if titulo in self.catalogo and self.catalogo[titulo].cantidad > 0:
            self.catalogo[titulo].cantidad -= 1
            self.pila_prestamo.append(titulo)
            print(f'Libro "{titulo}" prestado correctamente.')
        else:
            print(f'No hay disponibles instancias del libro "{titulo}".')

    def devolver_libro(self):
        if not self.pila_prestamo:
            print("No hay libros en préstamo para devolver.")
        else:
            print("Libros en préstamo:")
            for i, titulo in enumerate(reversed(self.pila_prestamo), 1):
                print(f"{i}. {titulo}")

            opcion = input("Seleccione el número del libro a devolver: ")
            try:
                opcion = int(opcion)
                if 1 <= opcion <= len(self.pila_prestamo):
                    titulo_devolver = self.pila_prestamo.pop(-opcion)
                    self.catalogo[titulo_devolver].cantidad += 1
                    print(f'Libro "{titulo_devolver}" devuelto correctamente.')
                else:
                    print("Número de opción no válido.")
            except ValueError:
                print("Por favor, ingrese un número válido.")

    def mostrar_estado(self):
        if not self.catalogo:
            print('La biblioteca no tiene libros en su catálogo.')
        else:
            print('Catálogo de la biblioteca:')
            for titulo, libro in self.catalogo.items():
                print(f'  - {titulo}: {libro.cantidad} disponibles')
            
            print('\nLibros en préstamo:')
            for titulo in reversed(self.pila_prestamo):
                print(f'  - {titulo}')


def main():
    biblioteca = Biblioteca()

    libro1 = Libro("Cien años de soledad", 3)
    libro2 = Libro("Don Quijote de la Mancha", 2)

    biblioteca.agregar_libro(libro1)
    biblioteca.agregar_libro(libro2)

    while True:
        print("\nOpciones:")
        print("1. Prestar libro")
        print("2. Devolver libro")
        print("3. Mostrar estado de la biblioteca")
        print("4. Salir")

        opcion = input("Seleccione una opción (1-4): ")

        if opcion == "1":
            titulo = input("Ingrese el nombre del libro a prestar: ")
            biblioteca.prestar_libro(titulo)
        elif opcion == "2":
            biblioteca.devolver_libro()
        elif opcion == "3":
            biblioteca.mostrar_estado()
        elif opcion == "4":
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, ingrese un número del 1 al 4.")


if __name__ == "__main__":
    main()
```

```python
'''Diseña un programa en Python que simule la atención de clientes en un banco. La atención
debe seguir el principio "primero en llegar, primero en ser atendido" (FIFO - First In, First
Out) utilizando una estructura de cola.'''

from collections import deque

class Banco:
    def __init__(self):
        self.cola_clientes = deque()

    def llegar_cliente(self, nombre):
        self.cola_clientes.append(nombre)
        print(f'{nombre} ha llegado al banco. Hay {len(self.cola_clientes)} personas en la cola.')

    def atender_cliente(self):
        if self.cola_clientes:
            cliente_atendido = self.cola_clientes.popleft()
            print(f'Atendiendo a {cliente_atendido}. Quedan {len(self.cola_clientes)} personas en la cola.')
        else:
            print('No hay clientes en la cola.')

def main():
    banco = Banco()

    while True:
        print("\nOpciones:")
        print("1. Llegar nuevo cliente")
        print("2. Atender cliente")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            nombre_cliente = input("Ingrese el nombre del nuevo cliente: ")
            banco.llegar_cliente(nombre_cliente)
        elif opcion == "2":
            banco.atender_cliente()
        elif opcion == "3":
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, ingrese un número del 1 al 3.")

if __name__ == "__main__":
    main()
```
